```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision 
import torch.optim as optim
from torchvision import datasets, transforms, models
from torchvision.models import resnet18
from torch.utils.data import DataLoader, WeightedRandomSampler
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
import numpy as np
from pylib import train_network
```

```python
#GPU 사용하여 학습 진행
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
print(device)

# ResNet18 모델 불러오기
model = resnet18(pretrained=True)
```

```python
input_channels = 1 #image 채널 수
num_classes = 4 #구분할 클래스 수
num_epoch = 100 #에폭 수 
batch_size = 32 #배치 사이즈 
learning_rate = 0.001
```

```python
# Define transformations (resize, normalize, etc.)
transform = transforms.Compose([
    transforms.Resize((224,224)),  # Resize to the required size
    transforms.Grayscale(num_output_channels=input_channels),  # Convert the image to 1 channel (grayscale)
    transforms.RandomResizedCrop(224, scale=(0.8, 1.0)),  # 무작위로 자르기 및 크기 조정
    transforms.RandomHorizontalFlip(),  # Horizontal flip
    transforms.RandomVerticalFlip(),  # Vertical flip
    transforms.ToTensor(),        # Convert the image to a tensor
 ])

# Load datasets from the folders
train_data = datasets.ImageFolder(root="./eye_data/train", transform=transform)
val_data = datasets.ImageFolder(root="./eye_data/val", transform=transform)
test_data = datasets.ImageFolder(root="./eye_data/test", transform=transform)

# 각 클래스의 샘플 개수를 세어 가중치를 계산
class_counts = [len(np.where(np.array(train_data.targets) == i)[0]) for i in range(len(train_data.classes))]
class_weights = [1.0 / count for count in class_counts]

# 학습 데이터셋의 각 샘플에 대한 가중치를 생성
sample_weights = [class_weights[label] for label in train_data.targets]

# 샘플러 정의
sampler = WeightedRandomSampler(weights=sample_weights, num_samples=len(sample_weights), replacement=True)

# Example DataLoader (adjust batch size as needed)
train_loader = DataLoader(train_data, batch_size=batch_size, sampler=sampler, num_workers=4)
val_loader = DataLoader(val_data, batch_size=batch_size, shuffle=False, num_workers=4)
test_loader = DataLoader(test_data, batch_size=batch_size, shuffle=False, num_workers=4)

# Example usage
x_example, y_example = train_data[0]
print(x_example.shape)  # 데이터 형상
print("Class Names:", train_data.classes)  # 클래스 이름 확인
print("Number of Classes:", len(train_data.classes))  # 클래스 수 확인
```

```python
#데이터셋 갯수 확인
print("train: ", len(train_data))
print("validation: ", len(val_data))
print("test: ", len(test_data))
```

```python
model.conv1 = nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)
'''
# FC 레이어 교체
num_features = model.fc.in_features  # 기존 FC 레이어의 입력 특징 수 추출
model.fc = nn.Sequential(
    nn.Linear(num_features, 128),  # 새로운 FC 레이어 추가
    nn.ReLU(),
    nn.Dropout(0.4),
    nn.Linear(128, 4)              # 최종 클래스 개수에 맞춤
)'''

# 모델을 디바이스로 이동
model.to(device)
```

```python
from torchsummary import summary

summary(model, (1, 224, 224))
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr = learning_rate)
```

```python
from torch.utils.tensorboard import SummaryWriter
writer = SummaryWriter('./runs/logs/')
```

```python
def images_to_tensorboard(writer, images, preds=None, step=0, name="Images"):
    img_grid = torchvision.utils.make_grid(images)
    writer.add_image(name, img_grid, global_step=step)
    if preds is not None:
        writer.add_text(f'{name}_Predictions', str(preds.tolist()), step)
```

```python
from tqdm import tqdm  # tqdm 라이브러리 추가

# Track the best model
best_val_acc = 0
best_model = None

# Training loop
for epoch in range(num_epoch):
    model.train()
    running_loss = 0.0
    correct = 0
    total = 0

    # 전체 배치 개수
    total_batches = len(train_loader)

    # tqdm을 이용한 진행 막대 표시
    with tqdm(total=total_batches, desc=f"Epoch {epoch+1}/{num_epoch}", unit="batch") as pbar:
        for i, (images, labels) in enumerate(train_loader):
            images = images.to(device)
            labels = labels.to(device)

            # Forward pass
            outputs = model(images)
            loss = criterion(outputs, labels)

            # Backward pass and optimization
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            # Track statistics
            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            # tqdm을 사용해 진행 막대 업데이트
            pbar.update(1)

            # Add a few random images to TensorBoard along with their labels (every 100 steps)
            if i % 100 == 0:
                images_to_tensorboard(writer, images, labels, step=epoch * len(train_loader) + i)

    train_acc = 100 * correct / total
    val_acc = 0
    val_loss = 0.0
    total = 0
    correct = 0

    # Validation step
    model.eval()
    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)
            loss = criterion(outputs, labels)
            val_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    val_acc = 100 * correct / total

    # Log to TensorBoard
    writer.add_scalar('Loss/train', running_loss / len(train_loader), epoch)
    writer.add_scalar('Accuracy/train', train_acc, epoch)
    writer.add_scalar('Loss/validation', val_loss / len(val_loader), epoch)
    writer.add_scalar('Accuracy/validation', val_acc, epoch)

    print(f'Epoch [{epoch+1}/{num_epoch}], Train Loss: {running_loss / len(train_loader):.4f}, '
          f'Train Acc: {train_acc:.2f}%, Val Acc: {val_acc:.2f}%')

    # Save the best model
    if val_acc > best_val_acc:
        best_val_acc = val_acc
        best_model = model.state_dict()

# Save the best model
torch.save(best_model, 'best_adam_resnet.pth')
writer.close()
```

```python
# 모델을 CPU로 로드
model.load_state_dict(torch.load('best_adam_resnet.pth'))
model.eval()
model.to('cpu')  # CPU 사용

correct = 0
total = 0
all_preds = []
all_labels = []
test_images = []

with torch.no_grad():
    for images, labels in test_loader:
        # 이미 CPU에서 처리하므로 추가적인 전송 필요 없음
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)

        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        # 예측값과 실제값을 리스트에 저장
        all_preds.extend(predicted.numpy())
        all_labels.extend(labels.numpy())
        test_images.append(images)  # 이미지도 CPU에 그대로 저장

# 테스트 정확도 계산
test_acc = 100 * correct / total
print(f'Test Accuracy: {test_acc:.2f}%')
```

```python
import matplotlib.pyplot as plt

def visualize_incorrect_predictions(images, labels, preds):
    # 틀린 예측 필터링
    incorrect_indices = [i for i, (true, pred) in enumerate(zip(labels, preds)) if true != pred]
    
    if incorrect_indices:
        incorrect_images = [images[i] for i in incorrect_indices]
        incorrect_labels = [labels[i] for i in incorrect_indices]
        incorrect_preds = [preds[i] for i in incorrect_indices]

        # 틀린 예측 시각화 (최대 16개 표시)
        fig, axes = plt.subplots(4, 4, figsize=(10, 10))  # 4x4 격자 형태
        axes = axes.flatten()
        for i, ax in enumerate(axes):
            if i < len(incorrect_images):
                img = incorrect_images[i].squeeze().numpy()
                ax.imshow(img, cmap='gray')
                ax.set_title(f'True: {incorrect_labels[i]}, Pred: {incorrect_preds[i]}')
                ax.axis('off')
            else:
                ax.axis('off')  # 남는 subplot 비우기
        plt.tight_layout()
        plt.show()
    else:
        print("모든 예측이 정확합니다!")

# 전체 데이터 준비
# test_images는 여러 배치로 나뉘어 있을 가능성이 있으므로 합칩니다.
all_test_images = torch.cat(test_images, dim=0)  # 모든 배치 연결
all_test_labels = all_labels  # 전체 레이블
all_test_preds = all_preds  # 전체 예측값

# 틀린 예측만 시각화
visualize_incorrect_predictions(all_test_images, all_test_labels, all_test_preds)
```

```python
from sklearn.metrics import confusion_matrix, f1_score, precision_score, recall_score, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

# Confusion Matrix 계산
cm = confusion_matrix(all_labels, all_preds)

# Precision, Recall, F1-Score 계산
precision = precision_score(all_labels, all_preds, average='weighted')
recall = recall_score(all_labels, all_preds, average='weighted')
f1 = f1_score(all_labels, all_preds, average='weighted')

# Confusion Matrix 시각화
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=sorted(set(all_preds)), yticklabels=sorted(set(all_labels)))
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

# 결과 출력
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1 Score: {f1:.2f}")

# 클래스별 상세 보고서
print("\nClassification Report:")
class_names = [f'Class {i}' for i in sorted(set(all_labels))]  # 클래스 이름 생성
print(classification_report(all_labels, all_preds, target_names=class_names))
```
